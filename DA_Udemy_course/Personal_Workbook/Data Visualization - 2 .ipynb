## Histogram - Introduction

```python
# Shows the distribution of a numeric variable
# The variable's range of values os split into intervals,represented by different bins
# The height of the bins shows the number of observations within an interval
# NOTE : (x,..,y) --> x and y are excluded
# [x,...,y] --> x and y are included
```

## Create a Histogram

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_real_estate = pd.read_csv('histogram_data.csv')
```

```python
df_real_estate
```

```python
# To plot histogram
plt.hist(df_real_estate['Price'])
plt.show()
```

```python
# Formatting
sns.set_style("white")
plt.figure(figsize= (8,6))

plt.hist(df_real_estate['Price'],
        bins= 8,
        color = '#108A99')

plt.title("Distribution of Real estate prices", fontsize = 14, weight = 'bold')
plt.xlabel("Price in (000 $)")
plt.ylabel("Number of Properties")
sns.despine()
plt.show()
```

```python
# We can notice form the above graph that it is a right skewed Distribution.
```

## How to make a good Histogram

```python
# There is usually NO definitive asnwer as to what the correct number of bins is.
# Scott's rule can be used when our data is continous or having a certain distribution
# Choosing a very large number of bins helps us see the data distribution
# The interval width can be deliberately altered.
# Without proper explanation , this might cause disinformation
```

```python
# When to use a Histogram?
# When we have numerical data (not necessarily)
```

## Scatter Plot - Intro

```python
# A scatter plot shows the relationship between two numerical features
# Features :
# 1. Can display a large number of points
# 2. Each observation is a point on that scatter
# 3. Our Data is not Aggregated
# 4. We can see the distinct values as well 

# Scatter plots is very frequently used in Data science
```

```python
# Why do we use scatter plots?
# 1. Can reveal relationships like correlation within the features
# 2. The concentration of ovservations and outliers in the data is easily visible 
# 3. A vital part of the exploratory data process for many statistical models
# 4. Easily recognize the outliers
```

```python
# A regression analysis would start by examining the relationships between the variables through a scatter plot or a correlogram.
# The scatter plot is often involved in the testing phases of modeling
```

## Creating Scatter plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_real_estate_scatter = pd.read_csv('scatter_data.csv')
```

```python
df_real_estate_scatter
```

```python
plt.figure(figsize= (9,6))
plt.scatter(df_real_estate_scatter['Area (ft.)'],
           df_real_estate_scatter['Price'])

plt.show()
```

```python
# To avoid overplotting we cvan add a transparency level
# Aplha : Controls the transparency level and takes vales from 0 to 1
# In scatter plot we can color code according to a third variable.

plt.figure(figsize= (9,6))
scatter = plt.scatter(df_real_estate_scatter['Area (ft.)'],
           df_real_estate_scatter['Price'],
           alpha= 0.6,   # ---> Adding alpha condition
           c= df_real_estate_scatter['Building Type'],  # --> add color coding
           cmap= 'viridis')  # --> add color coding
plt.legend(*scatter.legend_elements(),
          loc = "upper left",
          title = "Building Type")
plt.title("Relationship between Area and Price of California real estate",
         fontsize = 14,
         weight = "bold")
plt.xlabel ("Area (sq. ft.)", weight = "bold")
plt.ylabel("Price (000's of $)", weight = "bold")
plt.show()
```

```python
# Using seaborn
plt.figure(figsize= (9,6))
sns.scatterplot(df_real_estate_scatter['Area (ft.)'],
           df_real_estate_scatter['Price'],
               hue= df_real_estate_scatter['Building Type'],
               palette= ['black','darkblue','purple','pink','white'],
               s= 100)
plt.title("Relationship between Area and Price of California real estate",
         fontsize = 14,
         weight = "bold")
plt.xlabel ("Area (sq. ft.)", weight = "bold")
plt.ylabel("Price (000's of $)", weight = "bold")
plt.show()
```

```python
# Note : We were capable of plotting every observation from our dataset, based on its price and area
# Obersvation : Positive colinearity present between the two variables as a conclusion : 
# The larger the size of the property , the higher its price.
# Note  : one of the common mistakes is over plotting here
```

```python
# To avoid the overplotting extreme case : We could try to plot a subset of the entire data --> Sampling
# The sample should be random to avoid bais
# The sample shouold be representative of the pattern in data
```

## Create a Regression Scatter plot

```python
# All required lib. are already imported above
```

```python
df_ad_expenditure = pd.read_csv('scatter_plot_ii.csv')
```

```python
df_ad_expenditure
```

```python
# Sns.regplot() --> Regression plot
# sns.lmplot() --> Linear model plot
```

```python
# Regression plot
sns.regplot(x = "Budget",
           y = "Sales",
           data = df_ad_expenditure)
plt.show()
```

```python
#Formatting
sns.set(rc = {'figure.figsize' : (9,6)})
sns.regplot(x = "Budget",
           y = "Sales",
           data = df_ad_expenditure,
           color = 'grey',
           scatter_kws = {'color' :'k'},
           line_kws = {'color': 'red'})
plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's Units)")
plt.title("Effect of Ad Expenditure on Sales", fontsize = 14, weight = 'bold')
plt.show()
```

```python
# Linear Model plot
sns.lmplot(x = "Budget",
           y = "Sales",
           data = df_ad_expenditure)
plt.show()
```

```python
# Formatting 
sns.lmplot(x = "Budget",
           y = "Sales",
           data = df_ad_expenditure,
          height = 10,
           scatter_kws = {'color' :'k'},
           line_kws = {'color': 'red'})
plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's Units)")
plt.title("Effect of Ad Expenditure on Sales", fontsize = 14, weight = 'bold')
plt.show()
```

```python
# Note : The Great Benifit of the regression line is it's specificity
# A Regression line and regression equation helps us to quantify the exact relationship between the two
```

## Bar and Line chart(Combination Chart) Introduction

```python
# The combination chart is a hybrid of two or more chart types
# Most common example is Pareto chart where it is a combination of Line and Bar Chart
# In that Pareto line is showing the sum of relative frequencies(Cumulative Frequency)
# It's having dual Y axis where primary axis refers to bars and display the number of occurances
# Secondary Y axis refers to line chart and represents the Cumulative frequency
```

## Create a Bar and Line chart

```python
import pandas as pd
import matplotlib.pyplot as plt

# To format from numbers to percentage
from matplotlib.ticker import PercentFormatter
```

```python
df_kdnuggets = pd.read_csv('bar_line_chart_data.csv')
```

```python
df_kdnuggets
```

```python
# Here x-axis = Year
# Primary Y-axis = Participants
# Secondary Y-axis = Python Users
```

```python
# Create chart
fig, ax = plt.subplots(figsize =(10,7))

ax.bar(df_kdnuggets['Year'],
      df_kdnuggets['Participants'],
      color ='k')

ax1 = ax.twinx()
ax1.plot(df_kdnuggets['Year'],
        df_kdnuggets['Python Users'],
        color = '#b60000',
        marker = "D")
```

```python
# Formatting
########################## PRIMARY PLOT (BAR CHART)########################
fig, ax = plt.subplots(figsize =(10,7))

ax.bar(df_kdnuggets['Year'],
      df_kdnuggets['Participants'],
      color ='k')

ax.set_ylabel("Number of Participants",
             weight = 'bold')
ax.set_xlabel("Years",
             weight = 'bold')
ax.tick_params(axis = 'y',
              width = 2,
              labelsize = 'large')

ax.set_title("KD Nuggets Survey Python Users(2012 - 2019)", fontsize = 14, weight = 'bold')
######################## SECONDARY PLOT(LINE CHART) #######################
ax1 = ax.twinx()
ax1.set_ylim(0,1)  # -->Set the limit on Secondary Y Axis
ax1.yaxis.set_major_formatter(PercentFormatter(xmax = 1.0))

ax1.plot(df_kdnuggets['Year'],
        df_kdnuggets["Python Users"],
        color = '#b60000',
        marker = "D")

ax1.set_ylabel("Python Users in %",
              color = '#b60000',
             weight = 'bold')
ax1.tick_params(axis ='y',
               color = '#b60000',
               width = 2,
              labelsize = 'large')
```

```python
# NOTE : Combination chart is not frequently used as it is only for specific datasets.
# For better deliverables we can use two charts seperately for more understanding purpose.
# Bar - line and Scatter -line are the most popular charts in Data science field
```
