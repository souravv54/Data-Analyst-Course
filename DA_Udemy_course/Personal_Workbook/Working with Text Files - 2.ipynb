## Importing Data with NumPy .loadtxt() vs .genfromtxt()

```python
import numpy as np
```

```python
# np.loadtxt() : "load" implies the data is ready to be used after imported directly
# np.genfromtxt() : "generate" indicates that the function cretes the dataset from the text file
# Generating requires constructing the array as we go through the text file.
# loadtext is more faster whereas genfromtxt() is more flexible 
```

```python
lending_co_data_numeric_1 = np.loadtxt("Lending-Company-Numeric-Data.csv", delimiter= ',')
lending_co_data_numeric_1
```

```python
lending_co_data_numeric_2 = np.genfromtxt("Lending-Company-Numeric-Data.csv", delimiter= ',')
lending_co_data_numeric_2
```

```python
np.array_equal(lending_co_data_numeric_1,lending_co_data_numeric_2)
```

```python
lending_co_data_numeric_NaN = np.genfromtxt("Lending-Company-Numeric-Data-NAN.csv", delimiter= ';')
lending_co_data_numeric_NaN
# loadtxt is NOT working as NaN is not get handled from this function.
```

```python
#Hence we have converted all the values to string while using loadtxt function using dtype parameter
lending_co_data_numeric_NaN = np.loadtxt("Lending-Company-Numeric-Data-NAN.csv", delimiter= ';', dtype = np.str)
lending_co_data_numeric_NaN
```

```python
lending_co_data_numeric_NaN[0,0] + lending_co_data_numeric_NaN[0,1]
```

## Partially cleaning While importing - Numpy

```python
lending_co_data_numeric_NaN = np.genfromtxt("Lending-Company-Numeric-Data-NAN.csv", delimiter= ';')
lending_co_data_numeric_NaN
```

```python
# SKIP HEADER and FOOTER
lending_co_data_numeric_NaN = np.genfromtxt("Lending-Company-Numeric-Data-NAN.csv",
                                            delimiter= ';',
                                           skip_header= 2,
                                           skip_footer = 2)
lending_co_data_numeric_NaN
```

```python
# Fetch only columns
lending_co_data_numeric_NaN = np.genfromtxt("Lending-Company-Numeric-Data-NAN.csv", delimiter= ';',
                                           usecols= (0,1,5))
lending_co_data_numeric_NaN
```

```python
#Combine the clauses
lending_co_data_numeric_NaN = np.genfromtxt("Lending-Company-Numeric-Data-NAN.csv", delimiter= ';',
                                           usecols= (5,1,0),
                                           skip_header= 2,
                                           skip_footer = 2)
lending_co_data_numeric_NaN
```

```python
# Split the data into multiple variables 
lending_co_data_numeric_5,lending_co_data_numeric_0,lending_co_data_numeric_1 = np.genfromtxt("Lending-Company-Numeric-Data-NAN.csv", delimiter= ';',
                                                                                       usecols= (5,0,1),
                                                                                       skip_header= 2,
                                                                                       skip_footer = 2, unpack= True)
print(lending_co_data_numeric_5)
print(lending_co_data_numeric_0)
print(lending_co_data_numeric_1)
```

## Saving Your Data with Numpy - np.save()

```python
 import numpy as np
```

```python
lending_co = np.genfromtxt('Lending-Company-Saving.csv',
                            delimiter= ',',
                            dtype = np.str)

print(lending_co)
```

```python
np.save("Lending-Company-Saving", lending_co)
# It creates a .npy file in the same dir.
# NPY is a special type of text file native to Numpy
# NPY is much faster as compare to CSV
# It takes very less memory space.
# NPY is technically a text file
```

```python
# Load a dataset =/= import a dataset
# importing doesn't keep a track of the datatype of the original array
# We need to specify the datatype of the values after having brought into Python
# When loading, We don't need to specify or change our data while working with our python object.
```

```python
lending_data_save = np.load('Lending-Company-Saving.npy')
```

```python
print(lending_data_save)
```

```python
np.array_equal(lending_data_save,lending_co)
```

## Saving Your Data with Numpy - np.savez()

```python
import numpy as np
```

```python
lending_co = np.genfromtxt('Lending-Company-Saving.csv',
                            delimiter= ',',
                            dtype = np.str)

lending_data_save = np.load('Lending-Company-Saving.npy')
```

```python
np.savez('Lending-Company-Saving',lending_co)
```

```python
# np.savez() doesn't create an .npy but an . npz
# the NPZ is like an archive of NPYs that can store multiple arrays
# Instead of storing different datasets in seperate NPY files, we can store all of them in single NPZ
# it is convinent in large amount of data
```

```python
lending_data_savez = np.load('Lending-Company-Saving.npz')
```

```python
print(lending_data_savez)
```

```python
# This error come because NPZ contained collection of arrays
# It can't be properly displayed
```

```python
#Rectification
print(lending_data_savez["arr_0"])
# ["arr_0"] = [0]
```

```python
print(lending_data_savez["arr_1"])
```

```python
np.savez('Lending-Company-Saving', company = lending_co, data_save = lending_data_save)
```

```python
lending_data_savez = np.load("Lending-Company-Saving.npz")
```

```python
lending_data_savez.files
```

```python
print(lending_data_savez["company"])
```

```python
print(lending_data_savez["data_save"])
```

```python
np.array_equal(lending_data_savez["company"],lending_data_savez["data_save"])
```

## Saving Your Data with Numpy - np.savetxt()

```python
lending_co = np.genfromtxt('Lending-Company-Saving.csv',
                            delimiter= ',',
                            dtype = np.str)
```

```python
# np.savetxt() helps store numpy datasets in text files : .txt or .csv
# The syntax is once again very similar to np.save() and np.savez()
```

```python
np.savetxt("Lending-Company-Saving.txt",lending_co,fmt= '%s', delimiter=',')
```

```python
lending_data_savetxt = np.genfromtxt("Lending-Company-Saving.txt", delimiter= ',',dtype = np.str)
```

```python
print(lending_data_savetxt)
```

```python
lending_data_save = np.load("Lending-Company-Saving.npy")
```

```python
np.array_equal(lending_data_savetxt,lending_data_save)
```
