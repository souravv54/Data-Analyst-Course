## Create a pie chart in Python

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_fuel_engine_type = pd.read_csv('pie_chart_data.csv')
```

```python
df_fuel_engine_type
```

```python
# Creating a Pie Chart
plt.figure(figsize= (5,4))
plt.pie (df_fuel_engine_type['Number of Cars'],
        labels= df_fuel_engine_type['Engine Fuel Type'].values,
        autopct= '%.2f%%')
plt.show()
```

```python
# 'textdrops' is a dictionary of arguements for the text objects (for label formatting)
plt.figure(figsize= (5,4))
plt.pie (df_fuel_engine_type['Number of Cars'],
        labels= df_fuel_engine_type['Engine Fuel Type'].values,
        autopct= '%.2f%%',
        textprops= {'size':'large', 'fontweight' : 'bold','rotation' : '10'})
plt.show()
```

```python
# We will pick a colorblind friendly theme to make sure our chart is accessible to our entire audience.
```

```python
sns.set_palette('colorblind')
```

```python
plt.figure(figsize= (5,4))
plt.pie (df_fuel_engine_type['Number of Cars'],
        labels= df_fuel_engine_type['Engine Fuel Type'].values,
        autopct= '%.2f%%',
        textprops= {'size':'large', 'fontweight' : 'bold','rotation' : '10', 'color' : 'white'})
plt.show()
```

```python
# Add legend
plt.figure(figsize= (5,4))
plt.pie (df_fuel_engine_type['Number of Cars'],
        labels= df_fuel_engine_type['Engine Fuel Type'].values,
        autopct= '%.2f%%',
        textprops= {'size':'large', 'fontweight' : 'bold','rotation' : '10', 'color' : 'white'})
plt.legend()
plt.title('Cars by engine Fuel type', fontsize =15 , fontweight = 'bold')
plt.show()
```

## Pie Chart - Why Not to use OR avoid?

```python
# Pie chart can only be used for those cases where the data will get sumed up to 100%
# Generally Pie charts are been used to show the sub-parts of the whole
# Pie Chart are NOT recommended when :
# 1. You have the wrong type of data
# 2. Generally the pie slices are incomprehensive
# 3. When you have more than 5-6 Categories

# When using the 3-D effect, the proportions of a pie chart become even more Skewed
# We should also avoid doughnut pie charts
```

## Stacked Area Chart

```python
# Area Chart describes the Changes in Variable over time
# Both Area and line chart are a good option for timeline
# An area chart is used to display quantitative data for 3 or more numerical variables
# Classic area chart : Overlap among categories
# Stacked area chart : No overlap among categories (Entire area is sum of dataset)
# Steam graph : Displayed around a central category with NO overlap among categories
```

```python
# Features of Stacked area chart :
# 1. Conveys a lot of information.
# 2. Discovers trends and patterns in the data
# 3. Follows development of several different variables over time
```

## Create a Stacked Area chart

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_fuel_engine_types = pd.read_csv('stacked_area_chart_data.csv')
```

```python
df_fuel_engine_types.head(2)
```

```python
# Create a Stack Area chart
plt.figure(figsize= (12,6))
plt.stackplot(df_fuel_engine_types['Year'],
             df_fuel_engine_types['Diesel'],
              df_fuel_engine_types['Petrol'],
              df_fuel_engine_types['Gas'],
             )
plt.xticks(df_fuel_engine_types['Year'], rotation = 45)

plt.show()
```

```python
# Adding different color with color codes:
plt.figure(figsize= (12,6))
plt.stackplot(df_fuel_engine_types['Year'],
             df_fuel_engine_types['Diesel'],
              df_fuel_engine_types['Petrol'],
              df_fuel_engine_types['Gas'],
              colors= ["#011638","#7e2987","ef2026"],
              edgecolor = 'none')
plt.xticks(df_fuel_engine_types['Year'], rotation = 45)

plt.show()
```

```python
# Add lenged, labels and many more
labels = ["Disel","Petrol","Gas"]
sns.set_style("white")

plt.figure(figsize= (12,6))
plt.stackplot(df_fuel_engine_types['Year'],
             df_fuel_engine_types['Diesel'],
              df_fuel_engine_types['Petrol'],
              df_fuel_engine_types['Gas'],
              colors= ["#011638","#7e2987","ef2026"],
              edgecolor = 'none')
plt.xticks(df_fuel_engine_types['Year'], rotation = 45)
plt.legend(labels = labels , loc = 'upper left' )
plt.ylabel("Number of Cars", fontsize = 13)
plt.title("Popularity of Engine Fuel types(1982 - 2016)", fontsize = 14, weight = "bold")
sns.despine()  # Remove the border line 
plt.show()
```

## When is an area chart appropriate?

```python
# When we have time data
# When we want to compare the volume among categories
# When we have at least three categories
# When the variables are ordered and numerical
# When we have no more than 6-7 categories

# Important NOTE : Y-axis needs to start with Zero
```

## Create Line chart

```python
df_spx_ftse_00_10 = pd.read_csv('line_chart_data.csv')
```

```python
df_spx_ftse_00_10
```

```python
# Before we go to visualization , we need to correct the data format (Date columm)
df_spx_ftse_00_10.info()
```

```python
df_spx_ftse_00_10['Date'] = pd.to_datetime(df_spx_ftse_00_10['Date'])
```

```python
df_spx_ftse_00_10
```

```python
df_spx_ftse_00_10['Date'].info()
```

```python
plt.plot(df_spx_ftse_00_10['Date'],df_spx_ftse_00_10['GSPC500'])
plt.plot(df_spx_ftse_00_10['Date'],df_spx_ftse_00_10['FTSE100'])
plt.show()
```

```python
#Start Formatting
labels = ['GSPC500','FTSE100']

plt.figure(figsize= (28,8))
plt.plot(df_spx_ftse_00_10['Date'],df_spx_ftse_00_10['GSPC500'])
plt.plot(df_spx_ftse_00_10['Date'],df_spx_ftse_00_10['FTSE100'])
plt.title("S&P vs FTSE Returns(2000-2019)", fontsize = 14, fontweight = 'bold')
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels = labels, fontsize = 'large' )

plt.show()
```

```python
# We need to granularity of our chart
# We will chart Q3 and Q4 of 2008
df_spx_ftse_H2_08 = df_spx_ftse_00_10[(df_spx_ftse_00_10.Date >= '2008-07-01') &
                                        (df_spx_ftse_00_10.Date <= '2008-12-31')]
```

```python
df_spx_ftse_H2_08
```

```python
# New plot
labels = ['GSPC500','FTSE100']

plt.figure(figsize= (28,8))
plt.plot(df_spx_ftse_H2_08['Date'],df_spx_ftse_H2_08['GSPC500'], color = "midnightblue")
plt.plot(df_spx_ftse_H2_08['Date'],df_spx_ftse_H2_08['FTSE100'], color = "crimson")
plt.title("S&P vs FTSE Returns(H2 - 2008)", fontsize = 14, fontweight = 'bold')
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels = labels, fontsize = 'large' )

plt.show()
```

## When to use a Line Chart?

```python
# Line Chart are useful when :
# 1. We are displaying a time series data
# 2. We have several different variables
# 3. 3,5 or even 10 different variables over time
# 4. When we have multiple categories.

# A line chart is basically nothing but the tip of the area in the area chart
```
