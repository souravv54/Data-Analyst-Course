## 'Reason for Absence' - working with Dummy variables

```python
#Drop reason 0, avoid potential multicollinearity issues, as 0 is NOT making sense of having any reason for absence
```

```python
# Drop the first column from reason column dataframe
reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first= True)
reason_columns
```

## 'Reason for Absence' - Grouping 

```python
# Data frame columns
df.columns
```

```python
# Reason code columns
reason_columns.columns
```

```python
# If we need to insert the Reason codes , then it will bring up duplicates in terms of representing 
# the significance called multicollinearity .
# So to avoid that we can delete the column =' REASON FOR ABSENCE'
```

```python
# Step 1 : DROP
#Delete the column = Reason for Absence 
df = df.drop(['Reason for Absence'], axis = 1)
df
```

```python
#Step 2 : GROUPING
#Add all the column of reason codes (Reason columns) in dataframe,
# but we need to merge it by grouping the variable.
```

```python
# Re-organizing a certain type of variables into groups in a regression analysis.
```

```python
# We checked all the code reasons, and conclude that :
# Reason code 1-14 will come in group 1 as Diseases
# Reason code 15-17 will come into group 2 as pregnancy/child birth related reasons
# Reason code 18-21 will fall under group 3 as they are due to injury, poisioning etc etc.
# Reason code 22-28 will fall under group 4 as they come under light reasons like OPD,dental etc etc
```

```python
# After splitting this object into smaller pieces, each piece will consider as Dataframe
```

```python
reason_type_1 = reason_columns.loc[:,1:14].max(axis = 1)
reason_type_2 = reason_columns.loc[:,15:17].max(axis = 1)
reason_type_3 = reason_columns.loc[:,18:21].max(axis = 1)
reason_type_4 = reason_columns.loc[:,22:28].max(axis = 1)
```

```python
# Validation check 
reason_type_1
```

```python
# The above data is Pandas series or DataFrame?
# For now you can simply think as columns
# The structural distinction will be irrelevant when we get to the stage of attaching them to df
```

```python
# Step 3 : CONCATENATION 
df = pd.concat([df,reason_type_1,reason_type_2,reason_type_3,reason_type_4], axis = 1)
```

## Creating 'Day of the Week' from 'Date' Column
