## Q2. In which year highest number of TV shows & Movies were released ? show with Bar Graph

```python
data.head(2)
```

```python
data['Release_Date'] = pd.to_datetime(data.Release_Date)
```

```python
data['Date_N'] = data['Release_Date'].dt.year
```

```python
data.head(2)
```

```python
q1 = data.groupby('Date_N')['Year'].count()
```

```python
q1.plot(kind = 'bar')
```
